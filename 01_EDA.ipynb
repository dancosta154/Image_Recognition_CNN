### Wildfires by State

```python
df = pd.read_csv('./Data/wildfires_by_state_2021.csv')

# Remove unwanted states/countries
states_to_remove = ['Puerto Rico', 'United States (1)']
df_2021 = df[~df['state'].isin(states_to_remove)]

# Set index
df_2021 = df_2021.set_index('state')

# Convert floats to int
df_2021['num_fires'] = df_2021['num_fires'].astype(int)
df_2021['num_acres_burned'] = df_2021['num_acres_burned'].astype(int)

df_2021.head(15)
```

## Third Visual

Some basic cleaning of the tables occur here, followed by plotting a graph.

### Top 10 Costliest Wildfires in the United States (in millions)

```python
costliest_df = pd.read_csv('./Data/top_10_costliest_wf.csv')
```

```python
# Add State column and populate
state = ['CA', 'CA', 'CA', 'CA', 'CA', 'CA', 'CA', 'CA', 'CA', 'CA']
costliest_df['State'] = state
costliest_df
```

```python
# Rename columns
costliest_df.rename(columns={'rank': 'Rank', 
                             'year': 'Year', 
                             'name': 'Name', 
                             'dollars_when_occured': 'Dollars (Year Occurred)', 
                             'in_2021_dollars': 'Dollars (2021)'}, 
                    inplace=True)
costliest_df
```

```python
# Set index
costliest_df.set_index('Rank', inplace=True)
costliest_df
```

```python
# Plotting out the costliest fires, grouped by year

plt.figure(figsize=(12,6))
costliest_df.groupby('Year').sum()['Dollars (2021)'].sort_values(ascending=True).plot.barh(color = '#B92844')
plt.title('Costliest Wildfires in the U.S. by Year', fontsize=18)
plt.xlabel('Cost (in Millions)', fontsize=14)
plt.ylabel('Year', fontsize=14)
plt.savefig('./Images/costliest_wildfires_by_year.jpg', bbox_inches='tight', dpi=100); # save as jpg
```

```python
costliest_df.to_csv('./Data/costliest_wf_by_year.csv', index=True)
```
